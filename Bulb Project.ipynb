```python
import numpy as np
from persim import plot_diagrams
from ripser import ripser
from scipy.io import loadmat

# Function to calculate ERP for a single subject
def calculate_erp(subject_data):
    # Get unique smell IDs from the labels
    unique_smells = np.unique(subject_data['Label'])
    
    # Initialize dictionary to store ERPs for each channel and smell
    erp_data = {channel: {} for channel in ['FP1', 'AF7', 'AF3', 'F1', 'F3', 'F5', 'F7', 'FT7', 'FC5', 'FC3', 'FC1', 'C1', 
                 'C3', 'C5', 'T7', 'TP7', 'CP5', 'CP3', 'CP1', 'P1', 'P3', 'P5', 'P7', 'P9', 
                 'PO7', 'PO3', 'O1', 'Iz', 'Oz', 'POz', 'Pz', 'CPz', 'Fpz', 'Fp2', 'AF8', 'AF4', 
                 'AFz', 'Fz', 'F2', 'F4', 'F6', 'F8', 'FT8', 'FC6', 'FC4', 'FC2', 'FCz', 'Cz', 
                 'C2', 'C4', 'C6', 'T8', 'TP8', 'CP6', 'CP4', 'CP2', 'P2', 'P4', 'P6', 'P8', 
                 'P10', 'PO8', 'PO4', 'O2', 'Mstd_L', 'Mstd_R', 'BL1', 'BR1', 'BL2', 'BR2', 'BL3', 'BR3']}
    
    # Iterate over each unique smell
    for smell in unique_smells:
        # Get indices of trials corresponding to the current smell
        trial_indices = np.where(subject_data['Label'] == smell)[0]
        
        # Iterate over each channel
        for channel in erp_data.keys():
            # Extract signals for the current channel and smell
            signals = np.array([subject_data[channel][idx] for idx in trial_indices])
            
            # Calculate the ERP by averaging the signals
            erp_data[channel][smell] = np.mean(signals, axis=0)
    
    return erp_data

# Calculate ERPs for all subjects
all_subjects_erp = []

for subject_data in all_subjects_data:
    erp_data = calculate_erp(subject_data)
    all_subjects_erp.append({
        'Subject': subject_data['Subject'],
        'ERP': erp_data
    })
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
import pandas as pd

# Channel list
channels = ['FP1', 'AF7', 'AF3', 'F1', 'F3', 'F5', 'F7', 'FT7', 'FC5', 'FC3', 'FC1', 'C1', 'C3', 'C5', 'T7', 'TP7', 
            'CP5', 'CP3', 'CP1', 'P1', 'P3', 'P5', 'P7', 'P9', 'PO7', 'PO3', 'O1', 'Iz', 'Oz', 'POz', 'Pz', 'CPz',
            'Fpz', 'Fp2', 'AF8', 'AF4', 'AFz', 'Fz', 'F2', 'F4', 'F6', 'F8', 'FT8', 'FC6', 'FC4', 'FC2', 'FCz', 
            'Cz', 'C2', 'C4', 'C6', 'T8', 'TP8', 'CP6', 'CP4', 'CP2', 'P2', 'P4', 'P6', 'P8', 'P10', 'PO8', 
            'PO4', 'O2']

# Keys of interest
keys = [1, 2, 4, 8, 16, 32]

# Define lags to check for
lags_to_check = [1, 2, 3, 4]

# Initialize an empty list to store result DataFrames
results_list = []

# Initialize bl1 data
bl1 = all_subjects_erp[0]['ERP']['BL1']

# Preprocess bl1 (removing first 512 samples)
for key in bl1:
    bl1[key] = bl1[key][512:]

# Iterate over each channel
for channel in channels:
    # Fetch channel data and preprocess (removing first 512 samples)
    channel_data = all_subjects_erp[0]['ERP'][channel]
    for key in channel_data:
        channel_data[key] = channel_data[key][512:]
    
    # For each key, perform Granger causality test between channel and bl1
    for key in keys:
        # Prepare a DataFrame with the corresponding data
        df = pd.DataFrame(columns=[channel, 'bl1'], data=zip(channel_data[key], bl1[key]))
        
        # Perform Granger causality test for the defined lags
        gc_results = grangercausalitytests(df, 4)
        
        # Store p-values for each lag
        lag_p_values = {}
        for lag in lags_to_check:
            p_values = [
                gc_results[lag][0]['ssr_ftest'][1],  # p-value from ssr_ftest
                gc_results[lag][0]['ssr_chi2test'][1],  # p-value from ssr_chi2test
                gc_results[lag][0]['lrtest'][1],  # p-value from lrtest
                gc_results[lag][0]['params_ftest'][1],  # p-value from params_ftest
            ]
            if all(p < 0.05 for p in p_values):  # Check if all p-values are less than 0.05
                lag_p_values[lag] = p_values
        
        # If any lag satisfies the condition, add to the results list
        if lag_p_values:
            results_list.append(pd.DataFrame({
                'Channel': [channel],
                'Lag': [list(lag_p_values.keys())],  # Store the lags that satisfy the condition
                'Keys': [key],
                'P_Values': [list(lag_p_values.values())]  # Store the p-values
            }))

# Concatenate all the individual result DataFrames into one
if results_list:
    results_df = pd.concat(results_list, ignore_index=True)
else:
    results_df = pd.DataFrame(columns=['Channel', 'Lag', 'Keys', 'P_Values'])

# Display the final results DataFrame
print(results_df)
```

```python
from giotto.homology import VietorisRipsPersistence
from giotto.time_series import SingleTakensEmbedding
from giotto.diagrams import BettiCurve

# Initialize feature extraction tools
STE = SingleTakensEmbedding(parameters_type="search", n_jobs=-1, time_delay=10, dimension=5, stride=1)
VR = VietorisRipsPersistence(metric="euclidean", homology_dimensions=[1], n_jobs=-1, collapse_edges=True)
Bc = BettiCurve(n_bins=300)

# Initialize a list to store the data with features
all_subject_data_with_features = []

# Extract features for each subject and each channel-smell combination
for subject in all_subjects_erp:
    subject_name = subject['Subject']
    erp_data = subject['ERP']
    
    for channel in erp_data.keys():
        for smell_id, signal in erp_data[channel].items():
            # Apply Single Takens Embedding to the ERP signal
            y_embedded = STE.fit_transform(signal[1100:1850])  # Adjust indices as needed
            
            # Compute the persistence diagram using Vietoris-Rips complex
            diagrams = VR.fit_transform([y_embedded])
            
            # Extract Betti curve features from the persistence diagram
            features = Bc.fit_transform(diagrams)[0].T
            
            # Store the features along with the subject, channel, and smell_id
            all_subject_data_with_features.append({
                'subject': subject_name,
                'channel': channel,
                'smell_id': smell_id,
                'features': features
            })

# Print the result
for data in all_subject_data_with_features:
    print(f"Subject: {data['subject']}, Channel: {data['channel']}, Smell: {data['smell_id']}, Features shape: {data['features'].shape}")
```
